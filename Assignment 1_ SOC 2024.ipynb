```python
#data_fetcher.py

import yfinance as yf
import pandas as pd

def download_historical_data(symbol, start_date, end_date, timeframe='1d'):

    # Fetch data from Yahoo Finance
    data = yf.download(symbol, start=start_date, end=end_date, interval=timeframe)
    return data

```

```python
#performance.py

import matplotlib.pyplot as plt

def plot_stock_data(df, title="Stock Closing Prices", xlabel="Date", ylabel="Price"):

    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df['Close'], label='Closing Price')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#main.py

from datetime import datetime

# Define the parameters
symbol = 'RELIANCE.NS'
start_date = '2024-06-01'
end_date = datetime.today().strftime('%Y-%m-%d')

# Fetch the data
data = download_historical_data(symbol, start_date, end_date)
print(data.head())

# Plot the data
plot_stock_data(data, title=f'{symbol} Closing Prices from {start_date} to {end_date}')
```
